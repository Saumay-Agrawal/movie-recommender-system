```python
genre_set = set()
for genre in movies['genres'].unique():
    genre_set |= set(map(str, genre.split('|')))
genre_set = sorted(genre_set)
print(genre_set)
```

```python
movie_genre_matrix = np.zeros((movies.shape[0]+1, len(genre_set)+1), dtype=int)
counter = 0
for index, row in movies.iterrows():
    genres = list(map(str, row['genres'].split('|')))
    movie_genre_matrix[counter][0] = row['movieId']
    for i in range(len(genre_set)):
        if genre_set[i] in genres:
            movie_genre_matrix[counter][i+1] = 1
    counter += 1
movie_genre_matrix = pd.DataFrame(movie_genre_matrix, columns=['movieId']+genre_set)
movie_genre_matrix.head(10)
```

```python
def getMGvector(genres):
    v = []
    genres = genres.split('|')
    for genre in genre_set:
        if genre in genres:
            v.append(1)
        else:
            v.append(0)
    return v
```

```python
def getMGMatrix(df):
    mat = []
    for index, row in df.iterrows():
        temp = [row['movieId']] + getMGvector(row['genres'])
        mat.append(temp)
    cols = ['movieId'] + genre_set
    mat = pd.DataFrame(mat, columns=cols)
    mat.set_index('movieId', inplace=True)
    return mat
```

```python
mgmat = getMGMatrix(movies)
list(mgmat.loc[1, :])
```

```python
def getUGvector(df, mgmat):
    mat = []
    for index, row in df.iterrows():
        vec = list(mgmat.loc[row['movieId'], :])
        vec = [row['rating']*i for i in vec]
        mat.append(vec)
    n = len(mat)
    ugvec = []
    for i in range(20):
        gvalues = [mat[j][i] for j in range(n)]
        avg = round(sum(gvalues)/n, 3)
        ugvec.append(avg)
#     print(ugvec)
    return ugvec
```

```python
def getUGMatrix(df, mgmat):
    mat = []
    for user in df['userId'].unique():
        df1 = df.loc[df['userId']==user , ['movieId', 'rating']]
        df2 = mgmat.loc[df1['movieId'].values, :]
        temp = [user] + getUGvector(df1, df2)
        mat.append(temp)
    cols = ['userId'] + genre_set
    mat = pd.DataFrame(mat, columns=cols)
#     print(mat)
    mat.set_index('userId', inplace=True)
    return mat
```

```python
ugmat = getUGMatrix(ratings, mgmat)
ugmat.head(10)
```

```python
def getRecommendations(userid, ugm, mgm):
    mat = []
    ugv = ugm.loc[userid, :].values
    for movieid in mgm.index.values:
        mgv = mgm.loc[movieid, :].values
        score = round(np.dot(ugv,mgv), 3)
        mat.append([movieid, score])
    mat = pd.DataFrame(mat, columns=['movieId', 'score'])
    return mat
```

```python
scores = getRecommendations(1, ugmat, mgmat)
scores.sort_values(by=['score'], ascending=False, inplace=True)
for index, row in scores.head(10).iterrows():
    mid = row['movieId']
    mname = movies.loc[movies['movieId']==mid, 'title'].values[0]
    score = row['score']
    print('{}\t\t{}\t\t{}'.format(mid, mname, score))
```
