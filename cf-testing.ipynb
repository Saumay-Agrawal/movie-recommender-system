```python
def getSparsity(mat):
    density = float(len(mat.nonzero()[0]))/(mat.shape[0] * mat.shape[1])
    sparsity = 1-density
    sparsity *= 100
    print('Sparsity: {:4.2f}%'.format(sparsity))
```

```python
umm = pd.read_csv('./matrices/user-movie-matrix.csv')
umm.set_index('userId', inplace=True)
umm.head()
```

```python
getSparsity(umm.values)
print(umm.shape)
```

```python
usm = pd.read_csv('./matrices/user-similarity-matrix.csv')
usm.set_index('Unnamed: 0', inplace=True)
usm.head()
```

```python
getSparsity(usm.values)
print(usm.shape)
```

```python
msm = pd.read_csv('./matrices/movie-similarity-matrix.csv')
msm.set_index('Unnamed: 0', inplace=True)
msm.head()
```

```python
getSparsity(msm.values)
print(msm.shape)
```

```python
for i in msm.columns:
    if i not in umm.columns:
#         msm.drop(i, axis=0, inplace=True)
        msm.drop(i, axis=1, inplace=True)

getSparsity(msm.values)
print(msm.shape)
```

```python
def predictRatingUB(userid, movieid):
    similarity_vector = usm.iloc[:, userid]
    rating_vector = umm.iloc[:, movieid]
    rating = np.dot(similarity_vector, rating_vector)/(np.sum(np.abs(similarity_vector)))
    return round(rating*5, 2)
```

```python
predictRatingUB(2, 10)
```

```python
def predictRatingMB(userid, movieid):
    similarity_vector = msm.iloc[movieid, :]
    rating_vector = umm.iloc[userid, :]
    rating = np.dot(similarity_vector, rating_vector)/(np.sum(np.abs(similarity_vector)))
    return round(rating*5, 2)
```

```python
predictRatingMB(2, 10)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, n_init=20, random_state=0)
clusters = kmeans.fit_predict(umm.values)
```

```python
umm.index
```

```python
user_clusters = pd.DataFrame(clusters, index=umm.index, columns=['clusterId'])
print(user_clusters)
```

```python
def predict(userid, movieid):
    cid = user_clusters.loc[userid, 'clusterId']
    print(cid)
    sim_users = user_clusters.loc[(user_clusters['clusterId']==cid)]
    sim_users = list(sim_users.index)
    print(sim_users)
    similarity_vector = usm.iloc[sim_users, userid]
    rating_vector = umm.iloc[sim_users, movieid]
    rating = np.dot(similarity_vector, rating_vector)/(np.sum(np.abs(similarity_vector)))
    return round(rating*5, 2)
```

```python
predict(2,10)
```
